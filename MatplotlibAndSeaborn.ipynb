```python
years=range(2000,2012)
apples=[0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.939,0.941,0.943]
oranges=[0.962,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901,0.899,0.897,0.895]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.title('Crop Yield')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.legend(['Apples','Oranges'])
```

# markers

```python
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
plt.title('Crop Yield')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apples','Oranges']);
```

**Styling Lines and Markers**

```python
plt.plot(years,apples,marker='s',c='g',ls='-',lw=2,ms=8,mew=2)
plt.plot(years,oranges,marker='o',c='y',ls='--',lw=2,ms=10,mew=2,alpha=0.5)
plt.title('Crop Yield')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apples','Oranges']);
```

marker line color

format=[marker][line][color]

```python
plt.plot(years,apples,'s-g')
plt.plot(years,oranges,'o--r')
plt.title('Crop Yield')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apples','Oranges']);
```

**Changing The Figure Size**

```python
plt.figure(figsize=(6,4))
plt.plot(years,oranges,'s-y')
plt.title('crop yeild')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

# **Scatter Plot**

```python
data=sns.load_dataset("iris")
```

```python
data
```

```python
data.species.unique()
```

```python
plt.plot(data.sepal_length,data.sepal_width,'o-b');
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=data);
plt.figure(figsize=(8,5))
```

# Adding Hues

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',s=100,data=data);
plt.title("Iris dataset")
```

# Histogram

```python
data.sepal_width
```

```python
data.describe()
```

```python
plt.title("Distribution of sepal width")
sns.histplot(data.sepal_width);
```

```python
plt.title("Distribution of sepal width")
sns.histplot(data.sepal_width,bins=5);
```

```python
import numpy as np
plt.hist(data.sepal_width,bins=np.arange(2,5,0.25));
```

```python
plt.hist(data.sepal_width,bins=[1,3,4,4.5])
```

```python
setosa=data[data.species=='setosa']
versicolor=data[data.species=='versicolor']
virginica=data[data.species=='virginica']
```

```python
sns.histplot(setosa.sepal_width,bins=np.arange(2,5,0.25),alpha=0.4);
sns.histplot(versicolor.sepal_width,bins=np.arange(2,5,0.25),alpha=0.4);
sns.histplot(virginica.sepal_width,bins=np.arange(2,5,0.25),alpha=0.4);
plt.title("Distribution of sepal width");
plt.legend(['setosa','versicolor','virginica']);
```

```python
plt.hist([setosa.sepal_width,versicolor.sepal_width,virginica.sepal_width],bins=np.arange(2,5,0.25),stacked=True);
plt.title("Distribution of sepal width");
plt.legend(['setosa','versicolor','virginica']);
```

# **Heatmap**

```python
data2=sns.load_dataset("flights")
```

```python
data2
```

```python
plt.plot(data2.passengers)
```

```python
data21=data2.pivot(index="month",columns="year",values="passengers")
data21
```

```python
plt.title("No of Passengers (1000s)")
sns.heatmap(data21);
```

```python
plt.title("No of Passengers (1000s)")
sns.heatmap(data21,annot=True,cmap="Blues",fmt="d");
```
